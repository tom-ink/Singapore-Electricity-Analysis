```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files
# pd.set_option("display.max_columns", None)
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Data/Electricity Generation And Consumption, Annual, Gigawatt Hours (SingStat).csv')
df1a = df1.set_index(['index', 'subindex', 'subsubindex'])
df1a.columns = pd.to_datetime(df1a.columns).year
df1a
```

```python
df1a.xs('Electricity Generation').loc[:,'2005':].T.plot(kind='line', figsize=(16,12), color='red', marker ='o')

plt.title("Electricity Generation In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Generation (GWh)')
plt.ylim(30000, 60000)
plt.legend( ['Electricity Generation (GWh)']).remove()

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df1a.xs('Electricity Consumption').T.dropna(axis='index').plot(kind='bar', figsize=(16,12), stacked=True)

plt.title("Electricity Consumption in Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.ylim(0, 60000)
plt.legend(loc=(1.01, 0.4))
```

```python
df1a.xs('Electricity Consumption').groupby('subindex').sum().sort_values(by=2020).loc[:,2005:].T.dropna(axis='index').plot(kind='bar', figsize=(16,12), stacked=True)

plt.title("Electricity Consumption In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.legend(loc=(2), frameon=False)

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df1a.xs('Electricity Consumption').groupby('subindex').sum().sort_values(by=2020).loc[:,2005:].T.dropna(axis='index').plot(kind='line', figsize=(16,12), marker ='o')

plt.title("Overall Electricity Consumption In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.legend(loc=(2), frameon=False)

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df1a.xs('Industrial-related', level=1).loc[:,'2005':].T.plot(kind='line', figsize=(16,12), marker ='o', color=['blue', 'red', 'green', 'purple'])

plt.title("Industrial Related Electricity Consumption In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.legend(['Manufacturing', 'Construction', 'Utilities', 'Other Industrial-Related'],loc=(2), frameon=False)

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df1a.loc[('Electricity Consumption', 'Industrial-related', ['Construction', 'Utilities', 'Other Industrial-related']),'2005':].T.plot(kind='line', figsize=(16,12), marker ='o', color=['red', 'green', 'purple'])

plt.title("Industrial Related Electricity Consumption In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.legend(['Construction', 'Utilities', 'Other Industrial-Related'],loc=(2), frameon=False)

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df1a.xs('Commerce And Service-related', level=1).loc[:,'2005':].T.plot(kind='line', figsize=(16,12), marker ='o')

plt.title("Commerce & Service Related Electricity Consumption In Singapore")
plt.xlabel('Year')
plt.ylabel('Electricity Consumption (GWh)')
plt.ylim(0, 5500)
plt.legend(['Wholesale & Retail Trade', 'Accommodation & Food Services', 'Information & Communications',' Financial & Insurance Activities', 'Real Estate Activities', 'Professional, Scientific & Technical Activities', 'Other Commerce & Service'], loc=(2), frameon=False)

# plt.savefig("abc.png")
# files.download("abc.png")
```

```python
df2 = pd.read_csv('/content/drive/MyDrive/Data/Electricity Generation, Monthly, Gigawatt Hours (SingStat).csv')
df2a = df2.set_index('Unnamed: 0')
df2a.columns = pd.to_datetime(df2a.columns)
df2a.head()
```

```python
# Graph on monthly electricity generation in SG
df2a.T.plot(kind='line', figsize=(16, 10), legend=False)
```

```python
# Graph on electricity generation in SG by months
ax = df2a.T['2021-01-01':'2021-12-31'].plot(figsize=(12,8), legend=False, use_index=False)
df2a.T['2020-01-01':'2020-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
df2a.T['2019-01-01':'2019-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
df2a.T['2018-01-01':'2018-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
df2a.T['2017-01-01':'2017-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
df2a.T['2016-01-01':'2016-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
df2a.T['2015-01-01':'2015-12-31'].plot(figsize=(12,8), legend=False, ax=ax, use_index=False)
```

```python
# Percentage change in monthly electricity generation 
fig, ax = plt.subplots()

start_year = pd.date_range(start='1975-01-01', end='2021-12-01', freq='YS')
end_year = pd.date_range(start='1975-01-01', end='2021-12-31', freq='Y')
year_total = list(zip(start_year.to_list(), end_year.to_list()))

for y in year_total:
  df2a.T[y[0]:y[1]].pct_change().plot(figsize=(12,8), legend=False, ax=ax, use_index=False, linestyle='-', marker='.')

for y in year_total:
  df2a.T[y[0]:y[1]].pct_change().plot(figsize=(12,8), legend=False, ax=ax, use_index=False)

plt.axhline(y=0, color='red', linestyle='--', linewidth=1.5)
```
